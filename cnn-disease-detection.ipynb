<p class="report-content" style="line-height: 1.8; color: #333;">
We designed a CNN using Keras' Functional API with the following architecture:
</p>
<ol style="padding-left: 25px;">

</ol>
<p class="report-content" style="line-height: 1.8; color: #333;">

```python
# Set the path to your local dataset directory
dataset_dir = Path("C:/Users/dell/Downloads/PlantVillage/PlantVillage")
train_val_dir = Path('train_dataset')
test_dir = Path('test_dataset')

# Create temporary directories for train and test data
os.makedirs(train_val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

classes = os.listdir(dataset_dir)

# Split data into train and test
train_split = 0.7
test_split = 0.15

for class_name in classes:
    class_dir = dataset_dir / class_name
    images = list(class_dir.iterdir())
    np.random.shuffle(images)

    test_count = int(len(images) * test_split)

    train_val_images = images[:-test_count]
    test_images = images[-test_count:]

    os.makedirs(train_val_dir / class_name, exist_ok=True)
    os.makedirs(test_dir / class_name, exist_ok=True)

    for img in train_val_images:
        shutil.copy(img, train_val_dir / class_name)
    for img in test_images:
        shutil.copy(img, test_dir / class_name)
```

```python
# defining variables
BATCH_SIZE = 32
IMAGE_SIZE = 256
EPOCHS = 10
CHANNELS = 3
INPUT_SHAPE = (IMAGE_SIZE, IMAGE_SIZE, CHANNELS)
```

```python
datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    validation_split=0.15,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.4, 1.5],
    fill_mode="nearest",
)


test_data_generator = ImageDataGenerator(
    rescale=1.0 / 255,
)
```

```python
train_datagen = datagen.flow_from_directory(
    train_val_dir,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="training",
)

val_datagen = datagen.flow_from_directory(
    train_val_dir,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="validation",
)

test_datagen = test_data_generator.flow_from_directory(
    test_dir,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
)
```

```python
CLASS_NAMES = list(
    map(lambda x: str(x).split("___")[-1], train_datagen.class_indices.keys())
)
```

```python
def plot_random_batch(generator):
    """
    Plots a random batch of images from the given generator.
    """
    batch = next(generator)
    images, labels = batch

    cols = 6
    rows = BATCH_SIZE // cols + (BATCH_SIZE % cols > 0)

    fig, ax = plt.subplots(nrows=rows, ncols=cols, figsize=(11, 13))
    fig.tight_layout()
    fig.suptitle("Random Batch of Images", fontsize=20, y=1.04)
    fig.subplots_adjust(hspace=0.2, wspace=0.1)

    for i in range(BATCH_SIZE):
        row = i // cols
        col = i % cols
        ax[row, col].imshow(images[i])
        ax[row, col].axis("off")
        ax[row, col].set_title(CLASS_NAMES[np.argmax(labels[i])], fontsize=11)

    for j in range(BATCH_SIZE, rows * cols):
        fig.delaxes(ax.flatten()[j])

    plt.show()


# Plot a random batch
plot_random_batch(train_datagen)
```

    <p style="font-family: 'Arial', sans-serif; font-size: 16px; color: #555555; margin-bottom: 20px;">
        In this section, we define and compile the Convolutional Neural Network (CNN) model using Keras' Functional API. The model architecture begins with an input layer that takes images of shape (256, 256, 3). This is followed by a series of convolutional layers, each designed to capture spatial hierarchies in the image data. The first convolutional layer uses 32 filters of size 3x3, followed by a ReLU activation function to introduce non-linearity and a max-pooling layer to downsample the feature maps. This pattern is repeated with increasing filter sizes (64 and 128) to progressively capture more complex features.
    </p>
    <p style="font-family: 'Arial', sans-serif; font-size: 16px; color: #555555; margin-bottom: 20px;">
        After the convolutional and pooling layers, the feature maps are flattened into a single vector. This flattened vector is then fed into a dense (fully connected) layer with 128 neurons and a ReLU activation function, which helps in learning non-linear combinations of the high-level features extracted by the convolutional layers. The final layer is a dense layer with 4 neurons (corresponding to the number of output classes), using a softmax activation function to output a probability distribution over the classes.
    </p>
    <p style="font-family: 'Arial', sans-serif; font-size: 16px; color: #555555; margin-bottom: 20px;">
        Once the model architecture is defined, we compile the model. We use the Adam optimizer for its adaptive learning rate properties, and the categorical cross-entropy loss function, which is suitable for our multi-class classification problem. The accuracy metric is used to monitor the model's performance during training and validation.
    </p>
    <p style="font-family: 'Arial', sans-serif; font-size: 16px; color: #555555; margin-bottom: 20px;">
        The model is then trained for 20 epochs using the training data generator. To enhance the training process, we implement early stopping and model checkpointing. Early stopping is configured to monitor the validation loss and halt training if it does not improve for 3 consecutive epochs (patience), thus preventing overfitting and reducing unnecessary computation. The best model weights are restored after early stopping. Model checkpointing is set up to save the model at regular intervals (every 3 epochs) if there is an improvement in the validation loss, ensuring that we retain the best model version throughout the training process.
    </p>
</div>

```python
def image_classifier(input_shape, num_classes: int):
    inputs = Input(shape=input_shape)

    x = Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    x = MaxPooling2D((2, 2))(x)

    x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)

    x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)

    x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)

    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.1)(x)
    x = Dense(128, activation='relu')(x)
    outputs = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=inputs, outputs=outputs, name="potatoes-leave-clf-v1")
    return model
```

```python
model = image_classifier(INPUT_SHAPE, len(CLASS_NAMES))
model.summary(expand_nested=True)
```

```python
metrics = [
    keras.metrics.CategoricalAccuracy(name="accuracy"),
    keras.metrics.Precision(name="precision"),
    keras.metrics.Recall(name="recall"),
    keras.metrics.AUC(name="auc")
]

loss = keras.losses.CategoricalCrossentropy(from_logits=False)
optimizer = keras.optimizers.Adam()
EPOCHS = 20


early_stopping = EarlyStopping(
    monitor="val_loss", patience=3, restore_best_weights=True
)
checkpoint = ModelCheckpoint(
    "best_model.keras", monitor="val_loss", save_best_only=True, save_freq='epoch'
)

model.compile(
    loss=loss,
    optimizer=optimizer,
    metrics=metrics,
)


history = model.fit(
    train_datagen,
    epochs=EPOCHS,
    validation_data=val_datagen,
    callbacks=[checkpoint, early_stopping]
)
```

First, the plot_model_history function visualizes the training and validation accuracy and loss over epochs. It uses the model predictions and true labels from the data generators to calculate these metrics. Each dataset's metrics are visualized in a pie chart format, allowing for a quick comparison of the model's performance on each dataset.

The confusion matrix helps in understanding the model's classification performance by showing the count of true positive, true negative, false positive, and false negative predictions. It is accompanied by normalized values (percentages) within each class, providing a clear overview of the model's strengths and weaknesses in classification tasks.

```python
def plot_model_history(history):
    # Set the style
    sns.set_style("darkgrid")

    # Create color palette
    palette = sns.color_palette("viridis", 2)
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    def plot(ax, keys, labels, color1, color2, markers, title, legend_loc):
        ax.plot(
            history.history[keys[0]], label=labels[0], color=color1, marker=markers[0]
        )
        ax.plot(
            history.history[keys[1]], label=labels[1], color=color2, marker=markers[1]
        )
        ax.set_title(title, fontsize=16)
        ax.set_xlabel("Epoch", fontsize=14)
        ax.set_ylabel("Accuracy", fontsize=14)
        ax.legend(loc=legend_loc, fontsize=12)
        ax.grid(True)

    plot(
        axes[0],
        ("accuracy", "val_accuracy"),
        ("Training", "Validation"),
        palette[0],
        palette[1],
        ("o", "*"),
        "Model Accuracy",
        "lower right",
    )
    plot(
        axes[1],
        ("loss", "val_loss"),
        ("Training", "Validation"),
        palette[0],
        palette[1],
        ("o", "*"),
        "Model Loss",
        "upper right",
    )

    plt.tight_layout()
    plt.show()


plot_model_history(history)
```

```python
def plot_confusion_matrix(model, data_generator, num_samples, batch_size, class_names):
    data_generator.reset()

    ytrue = []
    ypred = []
    samples_processed = 0

    while samples_processed < num_samples:
        x_batch, y_batch = next(data_generator)
        batch_predictions = model.predict(x_batch, verbose=0)

        ypred.extend(np.argmax(batch_predictions, axis=1))
        ytrue.extend(np.argmax(y_batch, axis=1))

        samples_processed += len(y_batch)

    # Compute the confusion matrix
    cm = confusion_matrix(ytrue, ypred)
    cm_norm = cm / cm.sum(axis=1)[:, np.newaxis]

    annot = [
        f"{value}\n({pct:.1%})" for value, pct in zip(cm.flatten(), cm_norm.flatten())
    ]
    annot = np.array(annot).reshape(cm.shape)

    # Plot the confusion matrix
    plt.figure(figsize=(6, 4), dpi=100)
    sns.heatmap(
        cm,
        annot=annot,
        fmt="",
        cmap="Blues",
        xticklabels=class_names,
        yticklabels=class_names,
    )
    plt.xticks(size=10)
    plt.yticks(size=10)
    plt.title("Confusion Matrix - Test Dataset")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()


num_test_samples = test_datagen.n
plot_confusion_matrix(model, test_datagen, num_test_samples, BATCH_SIZE, CLASS_NAMES)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
)

def calculate_metrics(y_true, y_pred, y_pred_proba):
    """Calculate performance metrics."""
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average="weighted")
    recall = recall_score(y_true, y_pred, average="weighted")
    f1 = f1_score(y_true, y_pred, average="weighted")
    auc = roc_auc_score(y_true, y_pred_proba, multi_class="ovr", average="weighted")
    return accuracy, precision, recall, f1, auc

def evaluate_and_plot_metrics(model, train_gen, val_gen, test_gen, class_names):
    """Evaluate model on all datasets and plot metrics."""
    datasets = [("Train", train_gen), ("Validation", val_gen), ("Test", test_gen)]
    all_metrics = {}

    for name, generator in datasets:
        generator.reset()
        y_true, y_pred, y_pred_proba = [], [], []

        for x, y in generator:
            batch_pred = model.predict(x, verbose=0)
            y_true.extend(np.argmax(y, axis=1))
            y_pred.extend(np.argmax(batch_pred, axis=1))
            y_pred_proba.extend(batch_pred)

            if len(y_true) >= generator.n:
                break

        y_true = np.array(y_true)
        y_pred = np.array(y_pred)
        y_pred_proba = np.array(y_pred_proba)

        metrics = calculate_metrics(y_true, y_pred, y_pred_proba)
        all_metrics[name] = metrics

    # Plotting
    metric_names = ["Accuracy", "Precision", "Recall", "F1 Score", "AUC"]
    colors = ["#636EFA", "#EF553B", "#00CC96", "#AB63FA", "#FFA15A"]

    fig, axs = plt.subplots(1, 3, figsize=(18, 6), subplot_kw=dict(aspect="equal"))
    fig.suptitle("Model Performance Metrics", fontsize=24, y=1.05)

    for i, (name, metrics) in enumerate(all_metrics.items()):
        metrics_percent = [round(metric * 100, 2) for metric in metrics]
        def fmt(x):
            return '{:.2f}%'.format(sum(metrics_percent)*x/100)

        wedges, texts, autotexts = axs[i].pie(metrics_percent,
                                              labels=metric_names,
                                              colors=colors,
                                              autopct=fmt,
                                              startangle=90,
                                              pctdistance=0.75,
                                              wedgeprops=dict(width=0.5, edgecolor='white'))

        axs[i].set_title(name, fontsize=18)

        plt.setp(autotexts, size=11, weight="bold")

    plt.tight_layout()
    plt.show()

evaluate_and_plot_metrics(model, train_datagen, val_datagen, test_datagen, CLASS_NAMES)
```

The plot_feature_maps function is used to extract and display feature maps from specified layers of a trained model. These feature maps highlight the patterns and representations learned by the CNN as it processes input images.

```python
def plot_feature_maps(model, layer_names, input_image, max_images=20):
    outputs = [model.get_layer(name).output for name in layer_names]
    feature_map_model = Model(inputs=model.input, outputs=outputs)
    feature_maps = feature_map_model.predict(input_image)

    for fmap, layer_name in zip(feature_maps, layer_names):
        num_channels = fmap.shape[-1]

        num_images_to_plot = min(num_channels, max_images)

        fig = plt.figure(figsize=(num_images_to_plot * 2, 4))
        fig.suptitle(layer_name, y=0.78)
        fig.subplots_adjust(hspace=0.0, wspace=0.0)

        for i in range(num_images_to_plot):
            plt.subplot(1, num_images_to_plot, i + 1)
            plt.imshow(fmap[0, :, :, i], cmap="viridis")
            plt.axis("off")

        plt.show()
```
